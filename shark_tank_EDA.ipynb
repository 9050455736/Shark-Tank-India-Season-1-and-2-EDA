# Detection and handling missing values

```python
df.isnull().sum().to_frame().rename(columns={0:"missing values"}).T
```

### checking missing values in object data and numeric data separately

```python
df_obj=df.select_dtypes(include="object")
df_numeric=df.select_dtypes(exclude="object")
```

```python
ms_count=[]
ms_per=[]
dtypes=[]
cols=[]
for col in df_obj.columns:
    if df_obj[col].isnull().sum()>0:
        ms_count.append(df_obj[col].isnull().sum())
        ms_per.append(f"{df_obj[col].isnull().sum()*100/df.shape[0]}%")
        dtypes.append(df_obj[col].dtypes)
        cols.append(col)
ms_obj=pd.DataFrame({"columns":cols,"missing_value_count":ms_count,"missing_values_%":ms_per,"dtypes":dtypes})
ms_obj   
```

## checking missing values in numeric data

```python
ms_n=[]
ms_per=[]
col=[]
col_dtypes=[]
for i in df_numeric.columns:
    if df[i].isnull().sum()>0:
        col.append(i)
        ms_n.append(df[i].isnull().sum())
        ms_per.append(f"{df[i].isnull().sum()*100/df.shape[0]}%")
        col_dtypes.append(df[i].dtypes)
ms_numeric=pd.DataFrame({"columns":col,"missng_value_count":ms_n,"missng_value_count_%":ms_per,"dtypes":col_dtypes})
ms_numeric
```
